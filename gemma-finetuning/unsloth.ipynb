```python
import torch
from unsloth import FastLanguageModel

assert torch.cuda.is_available() and "A100" in torch.cuda.get_device_name(0), "Not on an A100"

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name    = "google/gemma-4-E2B",
    max_seq_length = 4096,
    load_in_4bit  = False,
    dtype         = torch.bfloat16,   # unsloth's kwarg is `dtype`; `torch_dtype` was silently ignored
)
FastLanguageModel.for_inference(model)

# `tokenizer` is a Gemma4Processor (multimodal) — arg 0 is `images`, not `text`.
# Use the inner text tokenizer for a raw continuation: no image slot, no chat template.
text_tok = getattr(tokenizer, "tokenizer", tokenizer)

prompt = "The process of photosynthesis in plants involves"
inputs = text_tok(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(**inputs, max_new_tokens=50, use_cache=True)
print("\n--- UNSLOTH OUTPUT ---")
print(text_tok.decode(outputs[0], skip_special_tokens=True))
```
